In the above 2-D figure we were not able to find which one is setosa, Virginia, or versicolor so we needed to identify the plot with help of seaborn we can do it.

```python
sns.set_style("whitegrid")
sns.scatterplot(data,x="sepal_length",y="sepal_width",hue="species")
plt.legend()
plt.show()
```

In observation, blue dots can be separated from other green and orange.
Because the green and orange are overlapping.
That we can use 3-D with the help of use Z factor the we can try to separated all three.

In this code, we can try to separate with using if the condition

```python
sep_len = float(input("Enter your sepal length: "))
sep_wid = float(input("Enter your sepal width: "))
if (3.5<sep_len<5.8) and (2.3<sep_wid<4.5):
  print("It is setosa")
elif (4.9<sep_len<7.0) and (2.0<sep_wid<3.4):
  print("It is versicolor")
elif (5.0<sep_len<7.9) and (2.3<sep_wid<3.8):
  print("It is virginica")
```

Calculating PDF Points

```python
data_setosa
data_versicolor
data_virgnica
counts,bin_edges = np.histogram(data_setosa['petal_length'], bins=10,density='True')
```

```python
np.histogram(data_setosa['petal_length'], bins=10,density='True')
```

```python
print("Printing the the counts of the histogram \n",counts)
pdf = counts/(sum(counts))
print("Sum of counts is \n",sum(counts))
print("Printing the bin of edges\n",bin_edges)
print("Pdf as follows\n",pdf)
plt.gca().legend(('Pdf'))
plt.title("PDF,PDF for iris setosa")
plt.xlabel("petal length")
plt.ylabel("Percentage")
plt.plot(bin_edges[1:],pdf)
plt.grid()
```

Calculating CDF points

```python
cdf = np.cumsum(pdf)
print("Print the cdf\n",cdf)
plt.gca().legend(('Cdf'))
plt.title('Cdf for iris setosa')
plt.xlabel('petal length')
plt.ylabel('Percentage')
plt.plot(bin_edges[1:],cdf)
plt.grid()
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.gca().legend(('Pdf','Cdf'))
plt.title('Pdf and Cdf for iris setosa')
plt.xlabel('petal length')
plt.ylabel('percentage')
plt.grid()
```

25% of points typically comes in between 1.5 to 1.6 is called a pdf

82% of points comes <= 1.6 called cdf

```python
counts,bin_edges = np.histogram(data_setosa['petal_length'], bins=10,density='True')
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges = np.histogram(data_setosa['petal_length'], bins=20,density='True')
pdf = counts/sum(counts)
plt.plot(bin_edges[1:],pdf)
plt.gca().legend(('Pdf','Cdf','bin edges'))
plt.title('Pdf and Cdf for iris setosa')
plt.xlabel('petal length')
plt.ylabel('percentage')
plt.grid()
```

PDF and CDF for all species

```python
#setosa
counts,bin_edges = np.histogram(data_setosa['petal_length'], bins=10,density='True')
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#versicolor
counts,bin_edges = np.histogram(data_versicolor['petal_length'], bins=10,density='True')
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#viriginica
counts,bin_edges = np.histogram(data_virgnica['petal_length'], bins=10,density='True')
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.title("PDF and CDF for all species")
plt.xlabel("Petal Length")
plt.ylabel("Percentage")
plt.grid()
```

```python
print("Means:")
print(np.mean(data_setosa['petal_length']))
print(np.mean(data_versicolor['petal_length']))
print(np.mean(data_virgnica['petal_length']))

print("Standard deviation:")
print(np.std(data_setosa['petal_length']))
print(np.std(data_versicolor['petal_length']))
print(np.std(data_virgnica['petal_length']))
```
