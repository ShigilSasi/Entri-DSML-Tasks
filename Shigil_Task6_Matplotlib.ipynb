##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
x = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
y = np.array([22.2,17.6,8.8,8,7.7,6.7])
plt.bar(x,y,color='Blue',edgecolor='Black',align='edge')
plt.title("Popularity of Programming Language \n"+"Worldwide, Oct2017 compared a year ago")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.minorticks_on()
plt.xlim(0,len(x))
plt.grid(which='major',linestyle='-',linewidth='0.5',color='Red')
plt.grid(which='minor',linestyle=':',linewidth='0.5',color='Black')
plt.show()
````

##Exercise 1: Read Total profit of all months and show it using a line plot

````python
import  seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('company_sales_data.csv.1')
sns.lineplot(x='month_number',y='total_profit',data=df)
plt.xticks(df['month_number'])
plt.yticks([100000,200000,300000,400000,500000])
plt.title("Company Profit Per Month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('company_sales_data.csv.1')
sales_data = (df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum())
mylabel = ["FaceCream","FaceWash","Toothpaste","BathingSoap","Shampoo","Moisturizer"]
plt.pie(sales_data,labels=mylabel,autopct='%1.1f%%')
plt.title("Sales Data")
plt.legend(loc='lower right')
plt.axis('equal')
plt.show()
````
