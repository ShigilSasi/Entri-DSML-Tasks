##4.  Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:



image.png

````python
import matplotlib.pyplot as plt
X1 = [10,20,30]
Y1 = [20,40,10]
X2 = [10,20,30]
Y2 = [40,10,30]
plt.axis([10,30,10,40])
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.plot(X1,Y1, color = 'blue', linewidth = 3, label = 'line1-width-3' )
plt.plot(X2,Y2, color = 'red', linewidth = 5, label = 'line2-width-5')
plt.title('Two or more lines with different widths and colors with suitable legends')
plt.legend()
plt.show()
````

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2,17.6,8.8,8,7.7,6.7]
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color='blue')
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Language\n' + 'Worldwide, Oct 2017 compared to a year ago')
plt.xticks(x_pos, x)
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':',linewidth='0.5', color='black')
plt.show()
````

##Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

image.png

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("company_sales_data.csv")

# Define the width of the bars
bar_width = 0.3

# Create an array of month numbers
months = data['month_number']

# Plot the bars for face cream
plt.bar(months - bar_width/2, data['facecream'], width=bar_width, label='Face Cream sales data')

# Plot the bars for face wash
plt.bar(months + bar_width/2, data['facewash'], width=bar_width, label='Face Wash sales data')

# Set x-axis ticks to show all months
plt.xticks(months)

# Add labels and legend
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales of Face Cream and Face Wash Over Months')
plt.legend(loc='upper left')

# Show the plot
plt.grid(linestyle='--')
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot

````python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('company_sales_data.csv')
x=data['month_number']

plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(x,data['facecream'],data['facewash'],data['toothpaste'],data['bathingsoap'],data['shampoo'],data['moisturizer'],
              colors=['m','c','r','k','g','y'])
plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')

plt.show()
````
